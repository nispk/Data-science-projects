# 911 Calls Capstone Project

For this capstone project we will be analyzing some 911 call data from [Kaggle](https://www.kaggle.com/mchirico/montcoalert). The data contains the following fields:

* lat : String variable, Latitude
* lng: String variable, Longitude
* desc: String variable, Description of the Emergency Call
* zip: String variable, Zipcode
* title: String variable, Title
* timeStamp: String variable, YYYY-MM-DD HH:MM:SS
* twp: String variable, Township
* addr: String variable, Address
* e: String variable, Dummy variable (always 1)

## Creating new features

** create a new column called "Reason" that contains the string value from title column.** 

**For example, if the title column value is EMS: BACK PAINS/INJURY , the Reason column value would be EMS. **

```python
df['Reason'] = df['title'].apply(lambda x :  
                 x.split(':')[0])

```

** the most common Reason for a 911 call based off of this new column **

```python
df['Reason'].value_counts()
```

** Now use seaborn to create a countplot of 911 calls by Reason. **

```python
import seaborn as sns
```

```python
sns.countplot('Reason', data=df)
```


** create 3 new columns called Hour, Month, and Day of Week. create these columns based off of the timeStamp column**

```python
df['Hour'] = df['timeStamp'].apply(lambda x: 
                                  pd.to_datetime(x).hour)
df['Month'] = df['timeStamp'].apply(lambda x:
                                   pd.to_datetime(x).month)
df['Day of Week'] = df['timeStamp'].apply(lambda x:
                                         pd.to_datetime(x).dayofweek).map({0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'})
```

** create a gropuby object called byMonth, where you group the DataFrame by the month column and use the count() method for aggregation.**

```python
byMonth = df.groupby('Month').count()
```

** create a simple plot off of the dataframe indicating the count of calls per month. **

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(byMonth.index ,byMonth['lat'], 'b')
ax.grid(True)
ax.set_xlabel('Month')
```

**  create a linear fit on the number of calls per month **

```python
byMonth.reset_index()
byMonth
```

```python
sns.lmplot(x ='Month', y = 'twp',data = byMonth.reset_index())
```

** Create a new column called 'Date' that contains the date from the timeStamp column ** 

```python
df['Date'] = df['timeStamp'].apply(lambda x:
                                  pd.to_datetime(x).date())
df
```

** groupby this Date column with the count() aggregate and create a plot of counts of 911 calls.**

```python
new_df = df.groupby('Date').count()
new_df.head()

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(new_df.index ,new_df['lat'], 'b')
ax.grid(True)
ax.set_xlabel('Date')
```

**  create 3 separate plots with each plot representing a Reason for the 911 call**

```python
newdf_Traffic = df[df['Reason'] == 'Traffic']
newdf_Fire = df[df['Reason'] == 'Fire']
newdf_EMS = df[df['Reason'] == 'EMS']
```

```python
newdf_date = newdf_Traffic.groupby('Date').count()
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(newdf_date.index, newdf_date['Reason'],'b')
ax.grid(True)

ax.set_title('Traffic')
```

```python
newdf_date = newdf_Fire.groupby('Date').count()
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(newdf_date.index, newdf_date['Reason'],'b')
ax.grid(True)

ax.set_title('Fire')
```

```python
newdf_date = newdf_EMS.groupby('Date').count()
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(newdf_date.index, newdf_date['Reason'],'b')
ax.grid(True)

ax.set_title('EMS')
```
